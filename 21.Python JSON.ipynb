```python
import json
```

```python
help(json)
```

```python
# string and object:----

x = {'name':"Trupti Vijay Kupekar","age":"26",'city':"Ghatkoper"}   # object
print(x)
print(type(x))
```

```python
x['name'] = "Trupti Kupekar"
print(x)
```

```python
x['name']
```

```python
# this is in string json and to convert into json object to perform modification
x = '{"name":"Trupti Vijay Kupekar","age":"26","city":"Ghatkoper"}'  # string
print(x)
print(type(x))
```

```python
# loads:----> it takes string data json and convert into json object.

y = json.loads(x)
print(y)
print(type(y))
```

```python
y['name']  = 'Trupti Kupekar'
print(y)
```

```python
print(type(y))
```

```python
#dumps: it takes object json and convert into string json:

print(y)
print(type(y))


# dumps

y2 = json.dumps(y)
print(y2)
print(type(y2))
```

```python
x = {
    'name':"Trupti",
    'age':26,
    'city':'New York'
}
print(x)
```

```python
# convert into json
y = json.dumps(x)
print(y)
```

# You can convert Python objects of the following types, into json string.

 * __dict__
 * __list__
 * __tuple__
 * __int__
 * __string__
 * __float__
 * __True__
 * __False__
 * __None__

```python
import json

print(json.dumps({'name':'John','age':45}))
print(json.dumps(['apple','banana']))
print(type(json.dumps(['apple','banana'])))
print(json.dumps(('apple','Mango','Papaya',"Elechie Banana")))
print(type(json.dumps(('apple','Mango','Papaya',"Elechie Banana"))))
print(json.dumps('Hello Trupti'))
print(json.dumps(100))
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

# When You convert from python to json, python objects are convert into json(javascript) equivalent

 * __Python__ : __JSON__
 * __dict__:    __Object__
 * __list__:    __Array__
 * __tuple__:   __Array__
 * __string__:  __String__
 * __integer__ : __Number__
 * __float__ : __Number__
 * __True__ : __true__
 * __False__: __false__
 * __None__ : __null__

```python
x = {
    'name':"John",
    "age": 26,
    'Height':5.8,
    'married': False,
    'divorced':False,
    'Childred':('Anmol',"Kriti"),
    "Pets":None,
    "Cars":[{"Model":"BMW 230","MPG":27.5},
           {'Model':"Ford Edge","mpg":24.1}
           ]
}
x
```

```python
print(type(x))
```

```python
str1 = json.dumps(x, indent=5)
print(str1)
print(type(str1))
```

```python
str1 = json.dumps(x, indent=5, separators = (".  ", " =  "))
print(str1)
print(type(str1))
```

```python
str1 = json.dumps(x, indent=5, sort_keys=False)
print(str1)
print(type(str1))
```

```python
str1 = json.dumps(x, indent=5, sort_keys=True)
print(str1)
print(type(str1))
```

# RegEx Functions

 * __Function__ : __Descriptions__
 * __findall__ : __Returns a list contatining all matches.__
 * __search__ : __Returns a Match object if there is a match anywhere in the string.__
 * __split__:__Returns a list where the string has been split at each match.__
 * __sub__: __Replaces one or many matches with a string.__

# Metacharacters

 * __Metacharacters are characters with special meaning:__

 * __Character__ : __Descriptions__:   __Example__
 * __[]__        : __""__:__A set of characters__:__"[a-m]"__
 * __\__         : __"Singnals a special sequence(can also be used to escape special characters)"__: __"\d"__
 * __.__         : __Any character (except newline character)__ : __"he.o"__

```python
import re
text = "The rain in Spain"

x = re.findall('ai',txt)
x
```

```python
text = "The rain in Spain"

x = re.findall('Trupti',txt)
x
```

```python
text = "The rain in Spain"

x = re.search('\sS',txt)
x
```

```python
# split() function

text = "The rain in Spain"

x = re.split('\s',txt)
x
```

```python
# sub()

text = "The rain in Spain"

x = re.sub("\s","-",txt)
x
```
