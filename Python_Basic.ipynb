```python
# Data typecasting in python

# Integer
print(100)
print(type(100))
print(1000)
print(type(1000))

# Float
print(45.60)
print(type(45.60))
print(45.0)
print(type(45.0))


# Boolean
print(True)
print(type(True))
print(False)
print(type(False))
```

```python
# Data Typecasting 

x =100
print(x)
print(type(x))

# float 100.0
y = float(x)
print(y)
print(type(y))

# sting
y1=str(x)
print(y1)
print(type(y1))
```

```python
# float

x1 =124.25

# integer
y1 = int(x1)
print(y1)
print(type(y1))

# string

x2 = str(x1)
print(x2)
print(type(x2))
```

```python
# Boolean

bool1 = True   # True =1
print(bool1)
print(type(bool1))


# integer
x3 = int(bool1)
print(x3)
print(type(x3))

# float
bool2= float(bool1)
print(bool2)
print(type(bool2))


# string
bool3=str(bool1)
print(bool3)
print(type(bool3))
```

```python
# False
false1 = False
print(false1)
print(type(false1))


# integer
false2 = int(false1)
print(false2)
print(type(false2))

#float
false3 = float(false1)
print(false3)
print(type(false3))
```
