# 18. Generators in Python

 * __Python generators are the function that returns transersal object and a simply way of creating iterators.__
 * __It transverse the entire items at once.__
 * __The generators can also be an expression in which syntax is similar to the list comprehension in python.__
 * __There is a lot of complixity in creating iteration in python, it is required to implement iter() and next() method to keep track of internal states.__
 * __It is a lengthy process to create iterators.__
 * __That is why the generator plays a significant role in simplyfying this process.__
 * __If there is no value found in iterations, it raises StopIteration exception.__
 * __It is quite simple to create a generator in Python.__
 * __It is similar to the normal function defined by the def keyword and employes a yield keyword instead of return.__
 * __The body of any function included a yield statement, it automatically become a generator functions.__
 * __The yield keyword is responsible to control the folw of the generator fuctions.__
 * __It pauses the functions execution when a successive function is called.__
 * __Later it resumes execution when a successive function is called.__
 * __The return keyword return a value and terminates the whole function and only one return statement can be employed in the functions.__

```python
def function():
    for i in range(25):
        if i%2==0:
            print(i)


function()
```

```python
def function():
    for i in range(25):
        if i%2==0:
            yield(i)

num_list=[]
for i in function():
    num_list.append(i)

print(num_list)
```

```python
def message_1():
    msg_one = "Hello world"
    yield msg_one

    msg_two = "Congratulation! Trupti"
    yield msg_two

    msg_three = "Python is the most popular porgramming language."
    yield msg_three

    msg_fourth = "Hello I am Trupti Vijay Kupekar"
    yield msg_fourth


obj1 = message_1()
print(next(obj1))
```

```python
def multi_table(n):
    for i in range(1,11):
        yield n*i
        i+=1

multi_table_list = []

for i in multi_table(19):
    multi_table_list.append(i)

print(multi_table_list)
```
